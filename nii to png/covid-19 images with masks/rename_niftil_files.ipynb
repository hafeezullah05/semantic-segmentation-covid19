```python
# import the modules
import os
from os import listdir
from natsort import natsorted
```

```python
# get the path or directory
dir_path = os.chdir("/Users/hafeez/Downloads/covid-19 images with masks/masks/coronacases-001/coronacases-001")

image_no = 1

for images in (natsorted(os.listdir(dir_path))):

    # check if the image end swith png or jpg or jpeg
	if (images.endswith(".png") or images.endswith(".jpg")\
		or images.endswith(".jpeg")):

            new_img_name = "image.{}.png".format(image_no)
            os.rename(images, new_img_name)

            #display images
            print(images,"converted to:",new_img_name)
    
            image_no = image_no+1
    

print("done re-naming all files")
```
